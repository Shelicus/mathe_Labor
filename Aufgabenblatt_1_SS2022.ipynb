a) Erstellen Sie eine Python-Funktion mit Namen F mit zwei Eingabevariablen (x, y), die auf die Eingabe von (0, 0) eine Eins zurückgibt. Für alle anderen Eingaben (x, y) $ \ne$ (0, 0) soll


$$\frac{sin(\sqrt{x{^2}+y{^2}})}{\sqrt{x{^2}+y{^2}}}$$

zurückgegeben werden. Testen Sie Ihre Funktion mit den Eingaben (0, 0),(0, 1),(5, 0) und (1, 1).


```python
def F(x,y):
    if x == 0 and y == 0:
        return 1

    else: 
        return((sin(sqrt(x**2 + y ** 2)))/sqrt(x**2 + y ** 2))
F(0,0)
F(0,1)       
F(5,0)    
F(1,1)    
```

c) Eine ganz andere Art, eine Funktion in zwei Variablen darzustellen, besteht darin, die eine Variable quasi als Zeit zu interpretieren. Erstellen Sie eine Animation der gleichen Funktion wie im Aufgabenteil a), indem Sie x als Ortsvariable und y als Zeitvariable auffassen. Sie bekommen dabei eine 2D– und nicht mehr eine 3D–Darstellung (Hinweis: FuncAnimation von dem matplotlib.animation). Stellen Sie die Animation auch für die Intervalle x = −15..15 und y = −15..15 dar.




```python
%matplotlib notebook 

fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], '-')
xdata = np.arange(-15, 15, 0.2)

def F(y):
    liste = []
    for i in xdata:
        if i == 0 and y == 0:
            liste.append(1)
        else:
            liste.append((sin(sqrt(i**2 + y ** 2)))/sqrt(i**2 + y ** 2))
    return liste

def init():
    ax.set_xlim(-15, 15)
    ax.set_ylim(-1, 1)
    ax.grid()
    return ln,

def update(frame):
    
    ydata = F(frame)
    ln.set_data(xdata, ydata)
    return ln,

step = 0.2
frames = np.arange(-15, 15, step)

ani = FuncAnimation(fig, update, frames, init_func=init, interval=100)

plt.show()
```
